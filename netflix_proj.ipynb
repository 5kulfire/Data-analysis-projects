```python
#PROJECT FOR DATA ANALYSIS BY RISHABH SINGH

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.express as px
from wordcloud import WordCloud 

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('netflix.csv')
df.head()
```

```python
#Renaming column
df.rename(columns={'listed_in':'genre'},inplace=True)
df.columns
```

```python
#Dropping show id
df.drop(columns='show_id',inplace=True)
```

```python
#Check missing values
df.isna().sum()
```

```python
#Dropping the null rows of date_added since there are only 10
df.dropna(subset=['date_added'],inplace=True)
df['date_added'].isna().sum()
```

```python
#Doing same for rating and duration columns
df.dropna(subset=['rating'],inplace=True)
df.dropna(subset=['duration'],inplace=True)
df.isna().sum()
```

```python
#Changing date_added to datetime format
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce') #coerce will ignore the errors if any
df.info()
```

```python
#Adding day,month,year,month_name,day_name as separate columns to help with analysis
df['day']=df['date_added'].dt.day
df['month']=df['date_added'].dt.month
df['year']=df['date_added'].dt.year
df['month_name']=df['date_added'].dt.month_name()
df['weekday']=df['date_added'].dt.day_name()
df.head()
```

```python
#Checking for unique ratings
df['rating'].unique()
```

```python
#Replacing rating into categories
df['rating']=df['rating'].replace(['TV-PG', 'TV-Y', 'TV-Y7','TV-G', 'G','TV-Y7-FV','PG'],'Kids')
df['rating']=df['rating'].replace(['PG-13','TV-14'],'Teens')
df['rating']=df['rating'].replace(['R','TV-MA','NC-17'],'Adults')
df['rating']=df['rating'].replace(['NR','UR'],np.NaN)
df.head()
```

```python
#Looking at null ratings
df['rating'].isna().sum()
```

```python
#Dropping the null ratings
df.dropna(subset=['rating'],inplace=True)
df['rating'].isna().sum()
```

```python
#Checking to see if there are any rows where all 3 columns-> director,country and cast are null
df[(df['director'].isna())&(df['country'].isna())&(df['cast'].isna())].shape[0]
df.shape
```

```python
#Dropping those 96 rows
df.dropna(subset=['director','cast','country'],how='all',inplace=True) #how='all' tells us that we have to drop everything inclusive in subset
df.shape
```

```python
#Working with cast column
df_cast=pd.DataFrame(df['cast'].apply(lambda x: str(x).split(', ')).tolist(),index=df['title'])
df_cast
```

```python
df_cast=df_cast.stack()
df_cast
```

```python
#changing into dataframe
df_cast=pd.DataFrame(df_cast)
df_cast
```

```python
df_cast.reset_index(inplace=True)
df_cast=df_cast[['title',0]]
df_cast.columns=['title','cast']
df_cast.replace('nan',np.nan,inplace=True)
df_cast.isna().sum()
```

```python
df_cast
```

```python
#Doing the same thing with genres
df_genre=pd.DataFrame(df['genre'].apply(lambda x: str(x).split(', ')).tolist(),index=df['title'])
df_genre=df_genre.stack()
df_genre=pd.DataFrame(df_genre)
df_genre.reset_index(inplace=True)
df_genre=df_genre[['title',0]]
df_genre.columns=['title','genre']
df_genre.replace('nan',np.nan,inplace=True)
df_genre.isna().sum()
```

```python
df_genre
```

```python
#Same for director column
df_dir=pd.DataFrame(df['director'].apply(lambda x: str(x).split(', ')).tolist(),index=df['title'])
df_dir=df_dir.stack()
df_dir=pd.DataFrame(df_dir)
df_dir.reset_index(inplace=True)
df_dir=df_dir[['title',0]]
df_dir.columns=['title','director']
df_dir.replace('nan',np.nan,inplace=True)
df_dir.isna().sum()
```

```python
df_dir
```

```python
#Same for country column
df_country=pd.DataFrame(df['country'].apply(lambda x: str(x).split(', ')).tolist(),index=df['title'])
df_country=df_country.stack()
df_country=pd.DataFrame(df_country)
df_country.reset_index(inplace=True)
df_country=df_country[['title',0]]
df_country.columns=['title','country']
df_country.replace('nan',np.nan,inplace=True)
df_country.isna().sum()
```

```python
df_country
```

```python
#merging cast and genre on the basis of title
df12=df_cast.merge(df_genre, on='title')
df12
```

```python
df123=df_dir.merge(df12,on='title')
df123
```

```python
df1234=df_country.merge(df123,on='title')
df1234
```

```python
#merging this df with the original df
df_new=df.merge(df1234,on='title')
df_new
```

```python
#dropping waste columns and renaming cleaned columns
df_new.drop(columns=['cast_x','country_x','director_x','genre_x'],inplace=True)
df_new.rename(columns={'cast_y':'cast','director_y':'director','genre_y':'genre','country_y':'country'},inplace=True)
df_new.head()
```

```python
df_new.isna().sum()
```

```python
df_new['cast']=df_new['cast'].replace(np.NaN,'Unknown')
df_new['country']=df_new['country'].replace(np.NaN,'Unknown')
df_new['director']=df_new['director'].replace(np.NaN,'Unknown')
df_new.isna().sum()
```

```python
df_new.dropna(inplace=True)
df_new.isna().sum()
```

```python
df_new.shape
```

```python
#Dropping duplicates
df_new.drop_duplicates(keep='first',inplace=True)
df_new.shape
```

```python
#Dividing the df on the basis of TV Shows and Movies
df_movies=df_new.loc[df_new['type']=='Movie']
df_TV=df_new.loc[df_new['type']=='TV Show']
```

```python
#replacing 'min' in Movie duration with empty string
df_movies['duration']=df_movies['duration'].str.replace('min','')
df_movies.head()
```

```python
#Doing same with TV Shows and replacing 'seasons'
df_TV['duration']=df_TV['duration'].str.replace('Seasons','')
df_TV['duration']=df_TV['duration'].str.replace('Season','')
df_TV.head()
```

```python
#Renaming duration to seasons for TV Shows
df_TV.rename(columns={'duration':'seasons'},inplace=True)
df_TV.head()
```

```python
#Plotting a pie chart
labels=['Movie','TV Show']
plt.pie(df_new.groupby('type')['type'].count(),labels=labels,explode=(0.08,0),autopct='%1.2f%%')
plt.suptitle('Movies vs TV Shows distribution')
plt.show()
```

```python
#Making a bar plot
df_temp=df_new.drop_duplicates(subset=['country','title']) #removing duplicates based on country and title

x=df_temp['country'].value_counts().head(10) #Top 10 countries

plt.figure(figsize=(17,8))
plt.title('Content across Countries')
label= sns.countplot(data=df_temp,x='country',hue='type',order=x.index,palette='magma')
for i in label.containers:
    label.bar_label(i) #this will show numerical value for each bar plot
plt.show()
```

```python
#Plotting Movies vs TV Shows based on directors
df_tv_temp=df_TV.drop_duplicates(subset=['director','title'])
df_movies_temp=df_movies.drop_duplicates(subset=['director','title'])

plt.figure(figsize=(17,7))
plt.suptitle('Top 10 Directors')

plt.subplot(1,2,1)
label=sns.countplot(data=df_tv_temp,x='director',order=df_tv_temp['director'].value_counts()[1:12].index,palette='plasma')
for i in label.containers:
    label.bar_label(i)
plt.title('Tv Shows')
plt.xticks(rotation=45)
plt.xlabel('Director')
plt.ylabel('Shows count')

plt.subplot(1,2,2)
label=sns.countplot(data=df_movies_temp,x='director',order=df_movies_temp['director'].value_counts()[1:12].index,palette='rocket')
for i in label.containers:
    label.bar_label(i)
plt.title('Movies')
plt.xticks(rotation=45)
plt.xlabel('Director')
plt.ylabel('Movies count')
plt.show()
```

```python
#Plotting Genres vs Show/Movie title based on directors
df_tv_temp=df_TV.drop_duplicates(subset=['genre','title'])
df_movies_temp=df_movies.drop_duplicates(subset=['genre','title'])

plt.figure(figsize=(17,7))
plt.suptitle('Top 10 Genres')

plt.subplot(1,2,1)
label=sns.countplot(data=df_tv_temp,x='genre',order=df_tv_temp['genre'].value_counts().head(10).index,palette='plasma')
for i in label.containers:
    label.bar_label(i)
plt.title('Tv Shows')
plt.xticks(rotation=45)
plt.xlabel('Genre')
plt.ylabel('Shows count')

plt.subplot(1,2,2)
label=sns.countplot(data=df_movies_temp,x='genre',order=df_movies_temp['genre'].value_counts().head(10).index,palette='rocket')
for i in label.containers:
    label.bar_label(i)
plt.title('Genre')
plt.xticks(rotation=45)
plt.xlabel('Genre')
plt.ylabel('Movies count')
plt.show()
```

```python
#Movie and TV SHows per country
df_temp=df_new.drop_duplicates(subset=['genre','title'])
plt.figure(figsize=(17,7))
plt.xticks(rotation=90)
sns.countplot(data=df_temp,x='country',hue='type',dodge=False,order=df_temp['country'].value_counts().head(30).index)
plt.show()
```

```python
#Pie chart based on content rating
df_movies_temp=df_movies.drop_duplicates(subset=['rating','title'])
df_tv_temp=df_TV.drop_duplicates(subset=['rating','title'])

plt.figure(figsize=(12,6))
plt.suptitle('Classification of content on Netflix')

plt.subplot(1,2,1) #Pie for movies
plt.pie(df_movies_temp['rating'].value_counts(),labels=df_movies_temp['rating'].value_counts().index,
        autopct='%.0f%%',explode=(0.03,0.02,0.02),shadow=True)
plt.title('Movies')

plt.subplot(1,2,2)
plt.pie(df_tv_temp['rating'].value_counts(),labels=df_tv_temp['rating'].value_counts().index,
        explode=(0.03,0.02,0.02),autopct='%.0f%%',shadow=True)
plt.title('TV Shows')

plt.show()
```

```python
#Movie releases by year, month and day 
df_movies_temp=df_movies.drop_duplicates(subset=['year','title'])

mv_year=df_movies_temp['year'].value_counts()
mv_year.sort_index(inplace=True)

month_order=['January','February','March','April','May','June','July','August','September','October','November','December']
mv_month=df_movies_temp['month_name'].value_counts().loc[month_order]

day_order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
mv_day=df_movies_temp['weekday'].value_counts().loc[day_order]
```

```python
#Plotting the graph

plt.figure(figsize=(17,8))
plt.suptitle('Movies added on Netflix')

plt.subplot(1,3,1) #Year wise grapg
label=sns.countplot(data=df_movies_temp,x='year',order=mv_year.index,palette='viridis')
for i in label.containers:
    label.bar_label(i)
plt.xticks(rotation=45)
plt.xlabel('Year')
plt.title('Year wise')

plt.subplot(1,3,2) #Month wise graph
label=sns.countplot(data=df_movies_temp,x='month_name',order=mv_month.index,palette='plasma')
for i in label.containers:
    label.bar_label(i)
plt.xticks(rotation=45)
plt.xlabel('Month')
plt.title('Month wise')

plt.subplot(1,3,3) #Weekday wise graph
label=sns.countplot(df_movies_temp,x='weekday',order=mv_day.index,palette='magma')
for i in label.containers:
    label.bar_label(i)
plt.xticks(rotation=45)
plt.xlabel('Weekday')
plt.title('Day wise')

plt.show()
```

```python
#TV Show releases by year, month and day 
df_tv_temp=df_TV.drop_duplicates(subset=['year','title'])

tv_year=df_tv_temp['year'].value_counts()
tv_year.sort_index(inplace=True)

month_order=['January','February','March','April','May','June','July','August','September','October','November','December']
tv_month=df_tv_temp['month_name'].value_counts().loc[month_order]

day_order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
tv_day=df_tv_temp['weekday'].value_counts().loc[day_order]

#Plotting the graph

plt.figure(figsize=(17,8))
plt.suptitle('TV Shows added on Netflix')

plt.subplot(1,3,1) #Year wise grapg
label=sns.countplot(data=df_tv_temp,x='year',order=tv_year.index,palette='winter')
for i in label.containers:
    label.bar_label(i)
plt.xticks(rotation=45)
plt.xlabel('Year')
plt.title('Year wise')

plt.subplot(1,3,2) #Month wise graph
label=sns.countplot(data=df_tv_temp,x='month_name',order=tv_month.index,palette='summer')
for i in label.containers:
    label.bar_label(i)
plt.xticks(rotation=45)
plt.xlabel('Month')
plt.title('Month wise')

plt.subplot(1,3,3) #Weekday wise graph
label=sns.countplot(df_tv_temp,x='weekday',order=tv_day.index,palette='spring')
for i in label.containers:
    label.bar_label(i)
plt.xticks(rotation=45)
plt.xlabel('Weekday')
plt.title('Day wise')

plt.show()
```
